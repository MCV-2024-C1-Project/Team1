```python
from similarity_calculator import SimilarityCalculator
from retrieval_system import RetrievalSystem
from evaluator import Evaluator
from data_loader import DataLoader
import pickle
imgs_ref = DataLoader({"dataset":"C:/Users/laila/Downloads/BBDD/BBDD"}).load_images_from_folder()
imgs_in = DataLoader({"dataset":"C:/Users/laila/Downloads/qsd1_w2/qsd1_w1"}).load_images_from_folder()
hist_ref = []
hist_query = []
# Get predictions
k_best_results = 1
for img in imgs_ref:
        height, width, channels = img.shape
        histogram_processor_1 = BlockBasedHistogram(img, block_size=(height//2,width//2), bins=8, normalize=True)
        hist_ref.append(histogram_processor_1.process_and_visualize(color_space='hsv', histogram_type='1D', combine_hs_ycrcb=True))
for img in imgs_in:
        height, width, channels = img.shape
        histogram_processor_1 = BlockBasedHistogram(img, block_size=(height//2,width//2), bins=8, normalize=True)
        hist_query.append(histogram_processor_1.process_and_visualize(color_space='hsv', histogram_type='1D', combine_hs_ycrcb=True))

    # Compute similarity
scores = SimilarityCalculator().compute_similarity(hist_query, hist_ref, method_name='MANHATTAN')

    # Get predictions
predictions = RetrievalSystem().retrieve_top_k(scores, reverse=False, k=k_best_results)
    # Load ground-truth
with open('C:/Users/laila/Downloads/qsd1_w2/qsd1_w1/gt_corresps.pkl', 'rb') as file:
    gt = pickle.load(file)

print(f"\nlen predictions: {len(predictions)}")
print(predictions)
print(f"\nlen gt: {len(gt)}")
print(gt)

    # Evaluate
mapk = Evaluator().mapk(gt, predictions, k_best_results)
print(f"mapk: {mapk}")
```
